```python
import pandas as pd

# Creating our dataset
downtime_data = {
    "Line": ["Line A", "Line B", "Line A", "Line B"],
    "Fault": ["Mechanical", "Electrical", "Electrical", "Mechanical"],
    "Hours_Lost": [2.5, 1.0, 4.0, 3.5]
}
df_downtime = pd.DataFrame(downtime_data)

# Grouping by "Line" to find the total hours lost for each
total_downtime = df_downtime.groupby("Line")["Hours_Lost"].sum()

print(total_downtime)
```

```python
import pandas as pd

# The Logistics Database
logistics_data = {
    "Supplier_ID": ["S-100", "S-101", "S-102"],
    "Material": ["Steel", "Aluminum", "Copper"],
    "Lead_Time_Days": [14, 7, 21]
}
df_logistics = pd.DataFrame(logistics_data)

# The Quality Control Database
qc_data = {
    "Supplier_ID": ["S-100", "S-101", "S-102"],
    "Defect_Rate": [1.2, 0.5, 4.3] # Percentages
}
df_quality = pd.DataFrame(qc_data)

combined_df = pd.merge(df_logistics, df_quality, on="Supplier_ID")
print(combined_df)
```

```python
import pandas as pd
import numpy as np

# System 1: Production Output Logs (Notice the missing data for Press-02)
production_data = {
    "Machine_ID": ["Press-01", "Press-02", "CNC-03", "Oven-04", "CNC-05"],
    "Department": ["Stamping", "Stamping", "Machining", "Thermal", "Machining"],
    "Units_Produced": [500, np.nan, 1200, 350, 950]
}
df_production = pd.DataFrame(production_data)

# System 2: Financial Cost Logs
cost_data = {
    "Machine_ID": ["Press-01", "Press-02", "CNC-03", "Oven-04", "CNC-05"],
    "Operating_Cost_USD": [1500, 1200, 3500, 4000, 2800]
}
df_costs = pd.DataFrame(cost_data)

df_production_clean = df_production.fillna(0)
df_factory = pd.merge(df_production_clean, df_costs, on="Machine_ID")
#print(df_factory)
df_factory["Cost_Per_Unit"] = df_factory["Operating_Cost_USD"] / df_factory["Units_Produced"]
a = df_factory.groupby("Department")[["Units_Produced", "Operating_Cost_USD"]].sum()
print(a)
```
