```python
# Example 1: Simple transformation
production = [120, 150, 80, 200, 95]

# Old way (traditional loop)
doubled = []
for p in production:
    doubled.append(p * 2)

# New Pythonic way (List Comprehension)
doubled_comprehension = [p * 2 for p in production]

print(doubled_comprehension)

# Example 2: With condition (filtering)
# Get only high production lines
high_production = [p for p in production if p > 100]
print("High production:", high_production)

# Example 3: Complex - with calculation
efficiencies = [78, 92, 65, 88, 95] 

# Create a list of "Good" or "Needs Improvement"
status = ["Good" if eff >= 85 else "Needs Improvement" for eff in efficiencies]
print(status)
```

```python
def production_planner(machines, efficiency, daily_output):
    """Simple Production Planning Tool"""
    print("=== Production Planning Simulator ===\n")
    
    # Calculate expected output
    expected_outputs = []
    for i in range(len(machines)):
        exp = daily_output[i] * (efficiency[i] / 100)
        expected_outputs.append(exp)
    
    # Print report
    total_expected = 0
    for i in range(len(machines)):
        print(f"{machines[i]:10} | Efficiency: {efficiency[i]:3}% | "
              f"Expected: {expected_outputs[i]:6.1f} units")
        total_expected += expected_outputs[i]
    
    print("\n" + "="*50)
    print(f"Total Expected Output from all machines: {total_expected:.1f} units")
    print("="*50)


# Test Data
machines = ["CNC-01", "Lathe-02", "Press-03", "Robot-04"]
efficiency = [92, 78, 88, 95]
daily_output = [450, 320, 520, 380]

production_planner(machines, efficiency, daily_output)
```

```python
aircraft_models = ["F-16", "F-22", "KAAN", "Eurofighter"]
fuel_levels = [45, 95, 100, 20]
needs_repairs = [True, False, False, False]

def check_fleet_status(models: list, fuels: list, repairs: list) -> None:
    # enumerate() counts for us, zip() bundles the lists together
    for hangar, (model, fuel, repair) in enumerate(zip(models, fuels, repairs), start=1):
        
        # Checking our two conditions
        if fuel > 80 and not repair:
            status = "Cleared for Takeoff"
        else:
            status = "Grounded"
            
        print(f"Hangar {hangar}: {model} is {status} (Fuel: {fuel}%)")

# Running the function
check_fleet_status(aircraft_models, fuel_levels, needs_repairs)
```

```python
laptop_models = ["Legion 5", "Omen Max 16", "ROG Zephyrus", "TUF Gaming"]
max_temps = [78, 92, 85, 75] # Peak temperatures in Celsius
has_vapor_chamber = [True, False, True, False]

def thermal_analysis(models: list, temps: list, chambers: list) -> None:
    for unit, (model, temp, chamber) in enumerate(zip(models, temps, chambers), start=1):
        if temp > 85 and not chamber:
            status = "Thermal Throttling Risk"
        else:
            status = "Cooling Adequate"

        print(f"Test Unit {unit}: {model} is {status} (Temperature is {temp}°C)")

thermal_analysis(laptop_models, max_temps, has_vapor_chamber)
```

```python
portfolio_stocks = ["NVDA", "AMD", "TTWO", "INTC"]
price_change = [4.2, -1.5, 2.1, -3.8]
high_volume = [True, False, False, True]

def market_scanner(stocks: list, changes: list, volumes: list) -> None:
    for alert, (stock, change, volume) in enumerate(zip(stocks, changes, volumes), start=1):
        if change < 0 and volume:
            status = "Active Monitoring Required"
        else:
            status = "Routine Volatility"
        
        print(f"Alert ID {alert}: {stock} is {status} (change is {change}%)")

market_scanner(portfolio_stocks, price_change, high_volume)
```

```python
components = ["Microcontrollers", "Sensors", "Displays", "Batteries"]
stock_counts = [50, 12, 100, 5]
unit_prices = [15, 40, 25, 60]

def audit_inventory(components: list, counts: list, prices: list) -> None:
    high_value_assets = []
    grand_total = 0
    for component, count, price in zip(components, counts, prices):
        line_value = count*price
        grand_total += line_value
        if line_value > 500:
            high_value_assets.append(component)

    print(f"Total ware house value {grand_total}$ and high value assets are {high_value_assets}")

audit_inventory(components, stock_counts, unit_prices)
```

```python
planned_output = [200, 150, 400, 300]
actual_output = [190, 150, 380, 285]

def calculate_efficiency(planned: list, actual: list) -> None:
    # blueprint: [(math expression) for (unpack variables) in zip(list1, list2)]
    efficiency_rates = [(a / p) * 100 for p, a in zip(planned, actual)]
    
    print(efficiency_rates)
calculate_efficiency(planned_output, actual_output)
```

```python
production_lines = ["F-22 Assembly", "KAAN Avionics", "Drone Chassis", "Engine Turbine"]
yield_rates = [98, 85, 99, 72] 
cycle_times = [150, 90, 45, 110] 

def factory_audit(lines: list, yields: list, times: list) -> None:
    # Part 1: The Formal Report
    # We unpack 3 variables from the zip to match our 3 lists
    for num, (name, y_rate, time) in enumerate(zip(lines, yields, times), start=1):
        if y_rate < 80:
            status = "Critical Rework Needed"
        else:
            status = "Optimal Production"
            
        print(f"Line {num}: {name} is {status} (Yield: {y_rate}%)")
        
    print("\n") 
    
    # Part 2: The Fast Filter
    # We use the exact same unpacking (name, y_rate, time) inside the list comprehension
    highly_efficient_lines = [name for name, y_rate, time in zip(lines, yields, times) if y_rate > 90 and time < 100]
    
    print(f"Highly Efficient Lines: {highly_efficient_lines}")

factory_audit(production_lines, yield_rates, cycle_times)
```

```python
trade_routes = ["Asia-Europe", "NA-Asia", "Europe-Africa", "SA-Europe"]
transit_days = [24, 18, 12, 30] 
delay_risks = [15, 5, 2, 25] # Percentage chance of delay

def route_analyzer(routes: list, days: list, risks: list) -> None:
    for num, (route, t_day, risk) in enumerate(zip(routes, days, risks), start=1):
        if risk > 20:
            status = "High Risk Route"
        else:
            status = "Standard Route"
        print(f"Route {num}: {route} is {status} (Risk: {risk}%)")
    print("\n") 

    fast_and_reliable = [route for route, t_day, risk in zip(routes, days, risks) if t_day < 20 and risk < 10]
    print(f"Fast and Reliable routes: {fast_and_reliable}")

route_analyzer(trade_routes, transit_days, delay_risks)
```
