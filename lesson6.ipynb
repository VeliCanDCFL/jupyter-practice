## 🧠 Concept: Cleaning Data and Saving Results

Real-world data is **never clean**. You'll encounter:
1. **Missing values** (NaN, null, empty)
2. **Duplicates** (same record twice)
3. **Need to save results** in different formats

---

## 📖 Part 1: Detecting Missing Data

Missing values appear as `NaN` (Not a Number) in pandas. Here's how to find them:

| Method | What it returns |
|--------|----------------|
| `.isna()` | True where value is missing |
| `.isnull()` | Same as `.isna()` |
| `.notna()` | True where value exists |
| `.notnull()` | Same as `.notna()` |

```python
# Find missing values in each column
df.isna().sum()          # Count of NaN per column

# Find rows with ANY missing value
df[df.isna().any(axis=1)]
```

---

## 📖 Part 2: Handling Missing Data

You have three main strategies:

| Strategy | Method | When to use |
|----------|--------|------------|
| **Drop** | `.dropna()` | Few missing values, data is random |
| **Fill** | `.fillna(value)` | Missing values have a sensible default |
| **Interpolate** | `.interpolate()` | Time series data (predict the gap) |

### Filling with Different Strategies

```python
# Fill with a constant
df.fillna(0)

# Fill different columns differently
df.fillna({'Age': 0, 'Name': 'Unknown', 'Salary': df['Salary'].mean()})

# Fill with column mean/median/mode
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

---

## 📖 Part 3: Removing Duplicates

```python
# Remove all duplicate rows
df.drop_duplicates()

# Remove duplicates based on specific columns
df.drop_duplicates(subset=['Email'])

# Count how many times each value appears
df['Email'].value_counts()

# How many unique values?
df['Email'].nunique()
```

---

## 📖 Part 4: Apply — Running Your Own Functions

`apply()` lets you run ANY function on your data.

| Scope | How to use |
|-------|-----------|
| **Series** (one column) | `df['col'].apply(func)` |
| **Row** (across columns) | `df.apply(func, axis=1)` |
| **DataFrame** (every cell) | `df.applymap(func)` |

```python
# Apply to each element in a column
df['Price'].apply(lambda x: x * 0.9)  # 10% off all prices

# Apply to each row
df.apply(lambda row: row['A'] + row['B'], axis=1)

# Apply to every cell in DataFrame
df.applymap(lambda x: round(x, 2))
```

---

## 📖 Part 5: Saving Data

| Method | Output Format |
|--------|--------------|
| `.to_csv()` | CSV (most universal) |
| `.to_excel()` | Excel (.xlsx) |
| `.to_json()` | JSON (for APIs/web) |
| `.to_parquet()` | Parquet (fast, compressed) |

```python
df.to_csv('output.csv', index=False)      # Don't save row numbers
df.to_excel('output.xlsx', index=False)
df.to_json('output.json', orient='records')
df.to_parquet('output.parquet', index=False)
```

````python
# ============================================
# MISSION 1: Detect Missing Data
# ============================================

import pandas as pd

df = pd.read_csv('sales_data.csv')
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")

# 1. Count NaN per column
print("=== NaN Count Per Column ===")
nan_counts = df.isna().sum()
print(nan_counts)

# 2. Verify with notna
print("\n=== Non-NaN Count Per Column (should add to total rows) ===")
not_nan = df.notna().sum()
print(not_nan)

# 3. Find rows with ANY missing value
print("\n=== Rows With Any Missing Value ===")
rows_with_na = df[df.isna().any(axis=1)]
print(f"Found {len(rows_with_na)} rows with at least one NaN")
print(rows_with_na[['CustomerName', 'Discount', 'Rating', 'City', 'Status']])
````

````python
# ============================================
# MISSION 2: Handle Missing Data
# ============================================

# 1. Fill Discount NaN with 0
df['Discount'] = df['Discount'].fillna(0)
print("=== Filled Discount NaN with 0 ===")
print(f"Remaining Discount NaN: {df['Discount'].isna().sum()}")

# 2. Fill Rating NaN with median
median_rating = df['Rating'].median()
df['Rating'] = df['Rating'].fillna(median_rating)
print(f"\n=== Filled Rating NaN with median ({median_rating}) ===")
print(f"Remaining Rating NaN: {df['Rating'].isna().sum()}")

# 3. Fill City NaN with 'Unknown'
df['City'] = df['City'].fillna('Unknown')
print(f"\n=== Filled City NaN with 'Unknown' ===")
print(f"Remaining City NaN: {df['City'].isna().sum()}")

# 4. Drop rows where Status is NaN
df_clean = df.dropna(subset=['Status'])
print(f"\n=== Dropped rows with NaN Status ===")
print(f"Rows before: {len(df)}, after: {len(df_clean)}")

# 5. Verify zero NaNs
print(f"\n=== Total NaN Remaining: {df_clean.isna().sum().sum()} ===")
````

````python
# ============================================
# MISSION 3: Remove Duplicates & Count Values
# ============================================

# 1. Check for duplicate TransactionIDs
print("=== Duplicate TransactionIDs ===")
dupes = df_clean['TransactionID'].duplicated().sum()
print(f"Duplicate count: {dupes}")

# 2. Remove duplicates
df_clean = df_clean.drop_duplicates(subset=['TransactionID'])
print(f"Rows after dedup: {len(df_clean)}")

# 3. Count orders per Category
print("\n=== Orders Per Category ===")
cat_counts = df_clean['Category'].value_counts()
print(cat_counts)

# 4. Number of unique cities
print(f"\n=== Unique Cities: {df_clean['City'].nunique()} ===")
print(df_clean['City'].nunique())
````

````python
# ============================================
# MISSION 4: Apply Custom Functions
# ============================================

# 1. Create a function for rating labels
def get_rating_label(rating):
    if rating >= 4:
        return 'Good'
    elif rating >= 3:
        return 'Average'
    else:
        return 'Poor'

# Apply to Rating column (Series)
df_clean['RatingLabel'] = df_clean['Rating'].apply(get_rating_label)
print("=== Rating Labels ===")
print(df_clean[['Product', 'Rating', 'RatingLabel']].head())

# 2. Row-wise function for Tax
df_clean['Tax'] = df_clean.apply(lambda row: round(row['TotalAmount'] * 0.08, 2), axis=1)
print("\n=== Tax (row-wise) ===")
print(df_clean[['Product', 'TotalAmount', 'Tax']].head())

# 3. Applymap to round all numeric columns
df_rounded = df_clean.applymap(lambda x: round(x, 2) if isinstance(x, float) else x)
print("\n=== Rounded DataFrame (first 3 rows) ===")
print(df_rounded.head(3))
````

````python
# ============================================
# MISSION 5: Save to All Formats
# ============================================

# 1. CSV
df_clean.to_csv('final_cleaned.csv', index=False)
print("✅ Saved to final_cleaned.csv")

# 2. Excel
df_clean.to_excel('final_cleaned.xlsx', index=False)
print("✅ Saved to final_cleaned.xlsx")

# 3. JSON
df_clean.to_json('final_cleaned.json', orient='records', indent=2)
print("✅ Saved to final_cleaned.json")

# 4. Parquet
df_clean.to_parquet('final_cleaned.parquet', index=False)
print("✅ Saved to final_cleaned.parquet")

# 5. Verify by reading back
df_verify = pd.read_csv('final_cleaned.csv')
print(f"\n=== Verification: {len(df_verify)} rows, {len(df_verify.columns)} columns ===")
````
