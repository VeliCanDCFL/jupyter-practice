## 🧠 Concept: How to Get Data Out of a DataFrame

Think of your DataFrame as a table. You need ways to:
1. **Pick specific columns** (get just the data you care about)
2. **Pick specific rows** (find the records that matter)
3. **Filter by conditions** (only show what matches your criteria)

---

## 📖 Part 1: Selecting Columns

### Single Column → Returns a Series

```python
df['ColumnName']        # Returns a Series (one column)
```

### Multiple Columns → Returns a DataFrame

```python
df[['Col1', 'Col2']]    # Returns a DataFrame (two columns)
# Note the DOUBLE brackets! This is a list inside a DataFrame.
```

---

## 📖 Part 2: Selecting Rows — `.loc[]` vs `.iloc[]`

This is the most confusing part for beginners. Here's the simple rule:

| Method | How it works | Example |
|--------|-------------|---------|
| `.loc[]` | **Label-based** — use column/row NAMES | `df.loc[0, 'Price']` → row label 0, column 'Price' |
| `.iloc[]` | **Integer-based** — use POSITION numbers | `df.iloc[0, 2]` → 1st row, 3rd column |

**Memory trick:**
- **loc = label** (both start with 'l')
- **iloc = integer location** (the 'i' stands for integer)

### `.loc[]` Examples

```python
df.loc[0]              # Row with label 0 (all columns)
df.loc[0, 'Price']     # Row 0, column 'Price' → single value
df.loc[0:2, ['Name', 'Price']]  # Rows 0-2, specific columns
```

### `.iloc[]` Examples

```python
df.iloc[0]             # First row (position 0)
df.iloc[0, 2]          # Row 0, column 2 → single value
df.iloc[0:3, 1:4]      # Rows 0-2, columns 1-3 (slicing like Python lists)
```

---

## 📖 Part 3: Filtering with Conditions

### Boolean Mask — The Core Concept

A **boolean mask** is just a Series of True/False values. When you pass it to `df[]`, pandas returns only the rows where the value is True.

```python
df[df['Price'] > 100]   # Only rows where Price > 100
```

### Multiple Conditions

| Symbol | Meaning | Python operator |
|--------|---------|----------------|
| `&` | AND | `and` (can't use 'and' in pandas!) |
| `\|` | OR | `or` (can't use 'or' in pandas!) |
| `~` | NOT | `not` (can't use 'not' in pandas!) |

**CRITICAL:** You MUST wrap each condition in parentheses:
```python
# ✅ Correct
df[(df['Category'] == 'Electronics') & (df['Price'] > 100)]

# ❌ Wrong — will cause an error!
df[df['Category'] == 'Electronics' & df['Price'] > 100]
```

---

## 📖 Part 4: Other Useful Filters

| Method | What it does | Example |
|--------|-------------|---------|
| `.query()` | SQL-like syntax | `df.query("Price > 100 and Category == 'Electronics'")` |
| `.isin()` | Check if value is in a list | `df[df['City'].isin(['NYC', 'LA'])]` |
| `.between()` | Range filter | `df[df['Age'].between(20, 30)]` |

## 💡 Key Takeaways

1. **`df['col']` = Series, `df[['col1','col2']]` = DataFrame** — double brackets for multiple columns!
2. **`.loc[]` uses labels/names, `.iloc[]` uses positions** — remember: loc=label, iloc=integer
3. **Boolean masks return True/False rows** — the core filtering mechanism
4. **Use `& | ~` not `and or not`** — and wrap each condition in parentheses!
5. **`.query()` is cleaner for complex filters** — reads like SQL

---

## ⚠️ Common Mistakes

| Mistake | Why it's wrong | Fix |
|---------|---------------|-----|
| `df[df['A']>5]['B']=0` | Chained indexing — may not work! | Use `df.loc[df['A']>5, 'B'] = 0` |
| Forgetting double brackets | `df[['col']]` not `df['col']` for multiple cols | Remember: list of columns needs list syntax |
| Using `and` instead of `&` | Python's `and` doesn't work element-wise | Use `&` with parentheses around each condition |

````python
# ============================================
# MISSION 1: Column Selection
# ============================================

import pandas as pd

# Load the data
df = pd.read_csv('sales_data.csv')
print(f"Loaded {len(df)} rows")

# 1. Single column → Series
print("=== Single Column (Series) ===")
products = df['Product']
print(type(products))  # <class 'pandas.core.series.Series'>
print(products.head())

# 2. Multiple columns → DataFrame (DOUBLE brackets!)
print("\n=== Multiple Columns (DataFrame) ===")
subset = df[['Product', 'Category', 'TotalAmount']]
print(type(subset))  # <class 'pandas.core.frame.DataFrame'>
print(subset.head())

# 3. All Electronics products' names and prices
print("\n=== Electronics Products ===")
electronics = df[df['Category'] == 'Electronics'][['Product', 'TotalAmount']]
print(electronics)
````

````python
# ============================================
# MISSION 2: .loc[] and .iloc[]
# ============================================

# Set TransactionID as index
df_loc = df.set_index('TransactionID')

# 1. .loc[5] — row with label 5, all columns
print("=== .loc[5] (row with label 5) ===")
print(df_loc.loc[5])

# 2. .iloc[0] — first row by position
print("\n=== .iloc[0] (first row by position) ===")
print(df.iloc[0])

# 3. .loc[1:5, ['Product', 'TotalAmount']] — rows 1-5, specific columns
print("\n=== .loc[1:5, ['Product', 'TotalAmount']] ===")
print(df_loc.loc[1:5, ['Product', 'TotalAmount']])

# 4. .iloc[0:5, 2:5] — first 5 rows, columns at positions 2-4
print("\n=== .iloc[0:5, 2:5] ===")
print(df.iloc[0:5, 2:5])
````

````python
# ============================================
# MISSION 3: Boolean Filtering
# ============================================

# Reset index for easier filtering
df_reset = df.reset_index()

# 1. Transactions over $500
print("=== TotalAmount > 500 ===")
expensive = df_reset[df_reset['TotalAmount'] > 500]
print(f"Found {len(expensive)} transactions")
print(expensive[['Product', 'TotalAmount']])

# 2. Electronics under $300 (use & for AND)
print("\n=== Electronics AND TotalAmount < 300 ===")
mid_electronics = df_reset[(df_reset['Category'] == 'Electronics') & (df_reset['TotalAmount'] < 300)]
print(f"Found {len(mid_electronics)} transactions")
print(mid_electronics[['Product', 'Category', 'TotalAmount']])

# 3. Books OR Clothing (use | for OR)
print("\n=== Books OR Clothing ===")
books_clothing = df_reset[(df_reset['Category'] == 'Books') | (df_reset['Category'] == 'Clothing')]
print(f"Found {len(books_clothing)} transactions")

# 4. Quantity between 2 and 4
print("\n=== Quantity between 2 and 4 ===")
mid_qty = df_reset[df_reset['Quantity'].between(2, 4)]
print(f"Found {len(mid_qty)} transactions")

# 5. Completed orders in West region
print("\n=== Completed AND West ===")
west_completed = df_reset[(df_reset['Status'] == 'Completed') & (df_reset['Region'] == 'West')]
print(f"Found {len(west_completed)} transactions")
````

````python
# ============================================
# MISSION 4: .query(), .isin(), .between()
# ============================================

# 1. .query() — SQL-like syntax
print("=== .query(): Books with TotalAmount > 100 ===")
books_expensive = df_reset.query("Category == 'Books' and TotalAmount > 100")
print(f"Found {len(books_expensive)} transactions")
print(books_expensive[['Product', 'TotalAmount']])

# 2. .isin() — check if in a list
print("\n=== .isin(): Electronics, Books, AND Clothing ===")
top_categories = df_reset[df_reset['Category'].isin(['Electronics', 'Books', 'Clothing'])]
print(f"Found {len(top_categories)} transactions")

# 3. .between() — range filter
print("\n=== .between(): Ratings between 4 and 5 ===")
high_rated = df_reset[df_reset['Rating'].between(4, 5)]
print(f"Found {len(high_rated)} transactions")
````

````python
# ============================================
# BONUS: Chained Operations
# ============================================

# Combine filtering + sorting + column selection in one chain
print("=== Chain: Electronics > $200, sorted by price (desc) ===")
result = (df_reset[df_reset['Category'] == 'Electronics']
          [df_reset['TotalAmount'] > 200]
          .sort_values('TotalAmount', ascending=False)
          [['Product', 'TotalAmount', 'City']])
print(result)
````
