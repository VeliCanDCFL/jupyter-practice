```python
# Machine efficiency example
machines = ['Machine A', 'Machine B', 'Machine C']
efficiency = [82, 91, 76]

plt.figure(figsize=(8, 5))
plt.bar(machines, efficiency, color=['skyblue', 'orange', 'green'])
plt.title('Machine Efficiency Comparison')
plt.ylabel('Efficiency (%)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Machine Efficiency Comparison - IE Example
import matplotlib.pyplot as plt

machines = ['Machine A', 'Machine B', 'Machine C']
efficiency = [82, 91, 76]

plt.figure(figsize=(8, 5))
bars = plt.bar(machines, efficiency, color=['skyblue', 'orange', 'green'])

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, 
             f'{height}%', ha='center', va='bottom', fontsize=11)

plt.title('Machine Efficiency Comparison', fontsize=14, pad=20)
plt.ylabel('Efficiency (%)')
plt.ylim(0, 100)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
