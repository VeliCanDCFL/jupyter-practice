# 📅 Day 1 — Creating & Inspecting Data
**Goal:** Use every creation and inspection function at least once.
**Dataset:** `sales_data.csv` (100 transactions)

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('sales_data.csv')
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
print(f"Columns: {list(df.columns)}")
```

## Task 1: Creating DataFrames from Scratch
Create DataFrames using every method below.

```python
# 1.1 Create from dict of lists
df_dict = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol'],
    'Age': [25, 30, 35],
    'City': ['NYC', 'LA', 'Chicago']
})
print("=== From Dict of Lists ===")
print(df_dict)
```

```python
# 1.2 Create from 2D list
df_list = pd.DataFrame(
    [[100, 'Laptop', 999.99], [200, 'Mouse', 45.00], [300, 'Keyboard', 75.00]],
    columns=['SKU', 'Product', 'Price']
)
print("\n=== From 2D List ===")
print(df_list)
```

```python
# 1.3 Create from dict using from_dict
data = {'Category': ['Electronics', 'Books', 'Clothing'],
        'Count': [45, 30, 25]}
df_from_dict = pd.DataFrame.from_dict(data)
print("\n=== From from_dict() ===")
print(df_from_dict)
```

```python
# 1.4 Create a Series
s = pd.Series([999.99, 45.00, 75.00], index=['Laptop', 'Mouse', 'Keyboard'], name='Price')
print("\n=== Series ===")
print(s)
print(f"Type: {type(s)}")
```

```python
# 1.5 Create date range
dates = pd.date_range('2024-01-01', periods=10, freq='D')
print("\n=== Date Range ===")
print(dates)

# Create a date-indexed Series
sales_series = pd.Series([100, 200, 150, 300, 250, 400, 350, 500, 450, 600], index=dates)
print("\n=== Date-indexed Series ===")
print(sales_series)
```

## Task 2: Reading Different File Formats
Read data from various sources.

```python
# 2.1 Read CSV (already done above)
print("=== read_csv() ===")
df_csv = pd.read_csv('sales_data.csv')
print(f"Rows: {len(df_csv)}, Columns: {len(df_csv.columns)}")
```

```python
# 2.2 Create and read Excel
df_to_excel = pd.DataFrame({'A': [1, 2, 3], 'B': ['x', 'y', 'z']})
df_to_excel.to_excel('temp_data.xlsx', index=False)
df_excel = pd.read_excel('temp_data.xlsx')
print("\n=== read_excel() ===")
print(df_excel)
```

```python
# 2.3 Create and read JSON
df_to_json = pd.DataFrame({'X': [10, 20], 'Y': ['a', 'b']})
df_to_json.to_json('temp_data.json')
df_json = pd.read_json('temp_data.json')
print("\n=== read_json() ===")
print(df_json)
```

## Task 3: Inspecting Data
Use every inspection function on the sales dataset.

```python
# 3.1 head() and tail()
print("=== First 5 rows (head) ===")
print(df.head())

print("\n=== Last 5 rows (tail) ===")
print(df.tail())
```

```python
# 3.2 info() — column dtypes and non-null counts
print("=== info() ===")
df.info()
```

```python
# 3.3 describe() — summary stats for numeric columns
print("=== describe() (numeric only) ===")
print(df.describe())
```

```python
# 3.4 describe(include='all') — summary for ALL columns
print("\n=== describe(include='all') ===")
print(df.describe(include='all'))
```

```python
# 3.5 shape — (rows, columns)
print(f"=== shape ===")
print(f"Shape: {df.shape}")
```

```python
# 3.6 columns — column names
print(f"\n=== columns ===")
print(df.columns.tolist())
```

```python
# 3.7 dtypes — data types of each column
print(f"\n=== dtypes ===")
print(df.dtypes)
```

```python
# 3.8 index — row index
print(f"\n=== index ===")
print(df.index)
```

```python
# 3.9 values — underlying numpy array
print(f"\n=== values (first 3 rows) ===")
print(df.values[:3])
```

```python
# 3.10 sample() — random sample
print("\n=== sample(n=5) ===")
print(df.sample(n=5))
```

```python
# 3.11 empty — check if DataFrame is empty
print(f"\n=== empty ===")
print(f"Is df empty? {df.empty}")
empty_df = pd.DataFrame()
print(f"Is empty_df empty? {empty_df.empty}")
```

## Task 4: Converting to Datetime
Convert the OrderDate column to datetime type.

```python
# pd.to_datetime()
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print("=== After converting OrderDate to datetime ===")
print(df.dtypes)
print(f"\nFirst order date: {df['OrderDate'].min()}")
print(f"Last order date: {df['OrderDate'].max()}")
```
