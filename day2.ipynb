# 📅 Day 2 — Selecting & Filtering Data
**Goal:** Use every selection and filtering function at least once.
**Dataset:** `sales_data.csv`

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## Task 1: Column Selection
Select single and multiple columns.

```python
# 1.1 Single column (returns Series)
print("=== df['Category'] ===")
print(type(df['Category']))
print(df['Category'].head())
```

```python
# 1.2 Multiple columns (returns DataFrame)
print("\n=== df[['Category', 'Product', 'TotalAmount']] ===")
print(type(df[['Category', 'Product', 'TotalAmount']]))
print(df[['Category', 'Product', 'TotalAmount']].head())
```

```python
# 1.3 .loc[] — label-based indexing
print("\n=== df.loc[row_label, col_label] ===")
# Set index first for meaningful loc usage
df_indexed = df.set_index('TransactionID')
print(df_indexed.loc[1, :])  # Row with ID=1, all columns
```

```python
# 1.4 .iloc[] — integer-position indexing
print("\n=== df.iloc[row_idx, col_idx] ===")
print(df_indexed.iloc[0, :])  # First row by position
```

```python
# 1.5 .at[] — single scalar access (fastest)
print("\n=== df.at[row_label, col_label] ===")
print(df_indexed.at[1, 'Category'])
```

```python
# 1.6 .iat[] — single scalar by position (fastest)
print("\n=== df.iat[row_idx, col_idx] ===")
print(df_indexed.iat[0, 2])  # First row, third column
```

## Task 2: Boolean Filtering
Filter rows using conditions.

```python
# 2.1 Simple boolean filter
print("=== df[df['TotalAmount'] > 500] ===")
expensive = df[df['TotalAmount'] > 500]
print(f"Found {len(expensive)} expensive transactions")
print(expensive[['Product', 'TotalAmount']].head())
```

```python
# 2.2 Multiple conditions with & | ~
print("\n=== Multiple conditions (Electronics AND TotalAmount > 300) ===")
filtered = df[(df['Category'] == 'Electronics') & (df['TotalAmount'] > 300)]
print(f"Found {len(filtered)} matching transactions")
print(filtered[['Product', 'Category', 'TotalAmount']])
```

```python
# 2.3 .query() — SQL-like syntax
print("\n=== df.query() ===")
queried = df.query("Category == 'Books' and TotalAmount > 100")
print(f"Found {len(queried)} book transactions over $100")
print(queried[['Product', 'TotalAmount']])
```

```python
# 2.4 .isin() — check if values are in a list
print("\n=== df['Category'].isin(['Electronics', 'Books']) ===")
selected = df[df['Category'].isin(['Electronics', 'Books'])]
print(f"Found {len(selected)} Electronics or Books transactions")
```

```python
# 2.5 .between() — range filter
print("\n=== df['Quantity'].between(2, 4) ===")
mid_qty = df[df['Quantity'].between(2, 4)]
print(f"Found {len(mid_qty)} transactions with quantity 2-4")
print(mid_qty[['Product', 'Quantity']].head())
```

## Task 3: Sorting
Sort by columns and index.

```python
# 3.1 sort_values() — single column ascending
print("=== sort_values('TotalAmount') (ascending) ===")
sorted_asc = df.sort_values('TotalAmount')
print(sorted_asc[['Product', 'TotalAmount']].head())
```

```python
# 3.2 sort_values() — single column descending
print("\n=== sort_values('TotalAmount', ascending=False) ===")
sorted_desc = df.sort_values('TotalAmount', ascending=False)
print(sorted_desc[['Product', 'TotalAmount']].head())
```

```python
# 3.3 sort_values() — multiple columns
print("\n=== sort_values(['Category', 'TotalAmount']) ===")
multi_sorted = df.sort_values(['Category', 'TotalAmount'], ascending=[True, False])
print(multi_sorted[['Category', 'Product', 'TotalAmount']].head(10))
```

```python
# 3.4 sort_index() — sort by index
print("\n=== sort_index() ===")
df_sorted_idx = df.set_index('CustomerName').sort_index()
print(df_sorted_idx.head())
```

## Task 4: Chained Operations
Combine selection, filtering, and sorting.

```python
# Chain: Filter → Sort → Select columns
print("=== Chain: Electronics > $200, sorted by price ===")
result = df[(df['Category'] == 'Electronics') & (df['TotalAmount'] > 200)] \
           .sort_values('TotalAmount', ascending=False) \
           [['Product', 'TotalAmount', 'City']]
print(result)
```

```python
# Chain with .loc[]
print("\n=== Using .loc[] for chained selection ===")
result2 = df.loc[df['Category'] == 'Books', ['Product', 'Quantity', 'TotalAmount']] \
            .sort_values('TotalAmount', ascending=False)
print(result2.head())
```
