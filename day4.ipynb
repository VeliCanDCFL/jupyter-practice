# 📅 Day 4 — Merging & Statistics
**Goal:** Use every merge and statistics function at least once.
**Dataset:** `sales_data.csv`

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## Task 1: Merging DataFrames
Create two small DFs and merge them with all join types.

```python
# Create two small DataFrames for merging
df1 = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Carol', 'David'],
    'City': ['NYC', 'LA', 'Chicago', 'Houston']
})

df2 = pd.DataFrame({
    'CustomerID': [2, 3, 4, 5],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Watch'],
    'Price': [999, 699, 450, 299]
})

print("=== df1 ===")
print(df1)
print("\n=== df2 ===")
print(df2)
```

```python
# 1.1 pd.merge() — inner join (default)
print("=== pd.merge(df1, df2, on='CustomerID') — INNER ===")
inner = pd.merge(df1, df2, on='CustomerID')
print(inner)
```

```python
# 1.2 pd.merge() — left join
print("\n=== pd.merge(..., how='left') ===")
left = pd.merge(df1, df2, on='CustomerID', how='left')
print(left)
```

```python
# 1.3 pd.merge() — right join
print("\n=== pd.merge(..., how='right') ===")
right = pd.merge(df1, df2, on='CustomerID', how='right')
print(right)
```

```python
# 1.4 pd.merge() — outer join
print("\n=== pd.merge(..., how='outer') ===")
outer = pd.merge(df1, df2, on='CustomerID', how='outer')
print(outer)
```

## Task 2: Concatenating DataFrames
Stack vertically and horizontally.

```python
# 2.1 pd.concat() — vertical (stack rows)
print("=== pd.concat([df1, df2]) — VERTICAL ===")
vertical = pd.concat([df1, df2], ignore_index=True)
print(vertical)
```

```python
# 2.2 pd.concat() — horizontal (stack columns)
print("\n=== pd.concat([df1, df2], axis=1) — HORIZONTAL ===")
horizontal = pd.concat([df1, df2], axis=1)
print(horizontal)
```

## Task 3: Statistics
Compute summary statistics on the sales dataset.

```python
# 3.1 .sum() — column-wise sum
print("=== df['TotalAmount'].sum() ===")
total_sales = df['TotalAmount'].sum()
print(f"Total sales: ${total_sales:,.2f}")
```

```python
# 3.2 .mean() — column-wise mean
print("\n=== df['TotalAmount'].mean() ===")
avg_sales = df['TotalAmount'].mean()
print(f"Average sale: ${avg_sales:,.2f}")
```

```python
# 3.3 .median() — column-wise median
print("\n=== df['TotalAmount'].median() ===")
median_sales = df['TotalAmount'].median()
print(f"Median sale: ${median_sales:,.2f}")
```

```python
# 3.4 .min() and .max()
print("\n=== .min() and .max() ===")
print(f"Min sale: ${df['TotalAmount'].min():,.2f}")
print(f"Max sale: ${df['TotalAmount'].max():,.2f}")
```

```python
# 3.5 .std() — standard deviation
print("\n=== df['TotalAmount'].std() ===")
std_sales = df['TotalAmount'].std()
print(f"Std dev: ${std_sales:,.2f}")
```

```python
# 3.6 .var() — variance
print("\n=== df['TotalAmount'].var() ===")
var_sales = df['TotalAmount'].var()
print(f"Variance: {var_sales:,.2f}")
```

```python
# 3.7 .count() — non-null count
print("\n=== df['TotalAmount'].count() ===")
count_sales = df['TotalAmount'].count()
print(f"Non-null sales: {count_sales}")
```

## Task 4: Advanced Math
Cumulative and difference operations.

```python
# 4.1 .cumsum() — cumulative sum
print("=== df['TotalAmount'].cumsum() (first 10) ===")
df_sorted = df.sort_values('OrderDate')
print(df_sorted[['Product', 'TotalAmount', 'TotalAmount']].head(10).assign(
    CumSum=df_sorted['TotalAmount'].cumsum()
))
```

```python
# 4.2 .cummax() — cumulative max
print("\n=== df['TotalAmount'].cummax() (first 10) ===")
result = df_sorted[['Product', 'TotalAmount']].assign(
    CumMax=df_sorted['TotalAmount'].cummax()
)
print(result.head(10))
```

```python
# 4.3 .cummin() — cumulative min
print("\n=== df['TotalAmount'].cummin() (first 10) ===")
result = df_sorted[['Product', 'TotalAmount']].assign(
    CumMin=df_sorted['TotalAmount'].cummin()
)
print(result.head(10))
```

```python
# 4.4 .diff() — difference between consecutive rows
print("\n=== df['TotalAmount'].diff() (first 10) ===")
result = df_sorted[['Product', 'TotalAmount']].assign(
    Diff=df_sorted['TotalAmount'].diff()
)
print(result.head(10))
```

```python
# 4.5 .pct_change() — percentage change
print("\n=== df['TotalAmount'].pct_change() (first 10) ===")
result = df_sorted[['Product', 'TotalAmount']].assign(
    PctChange=df_sorted['TotalAmount'].pct_change()
)
print(result.head(10))
```
