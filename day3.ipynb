# 📅 Day 3 — Modifying & GroupBy
**Goal:** Use every modification and groupby function at least once.
**Dataset:** `sales_data.csv`

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## Task 1: Adding/Removing Columns
Add new columns, drop existing ones.

```python
# 1.1 Add new column with assignment
print("=== df['DiscountAmount'] = ... ===")
df['DiscountAmount'] = df['TotalAmount'] * df['Discount'].fillna(0)
print(df[['Product', 'TotalAmount', 'Discount', 'DiscountAmount']].head())
```

```python
# 1.2 Add another column
df['PriceAfterDiscount'] = df['TotalAmount'] - df['DiscountAmount']
print("\n=== PriceAfterDiscount ===")
print(df[['Product', 'TotalAmount', 'DiscountAmount', 'PriceAfterDiscount']].head())
```

```python
# 1.3 .drop() — drop column
print("\n=== df.drop('DiscountAmount', axis=1) ===")
df_dropped = df.drop('DiscountAmount', axis=1)
print(f"Columns before: {len(df.columns)}, after: {len(df_dropped.columns)}")
```

```python
# 1.4 .drop() — drop rows by index
print("\n=== df.drop([0, 1, 2], axis=0) ===")
df_rows_dropped = df.drop([0, 1, 2], axis=0)
print(f"Rows before: {len(df)}, after: {len(df_rows_dropped)}")
```

```python
# 1.5 .rename() — rename columns
print("\n=== df.rename(columns={...}) ===")
df_renamed = df.rename(columns={'CustomerName': 'Client', 'TotalAmount': 'Revenue'})
print(df_renamed.columns.tolist())
```

```python
# 1.6 .set_index() — set column as index
print("\n=== df.set_index('TransactionID') ===")
df_indexed = df.set_index('TransactionID')
print(df_indexed.head())
```

```python
# 1.7 .reset_index() — reset index to default
print("\n=== df.reset_index() ===")
df_reset = df_indexed.reset_index()
print(df_reset.head())
```

## Task 2: Advanced Modification
Use assign and astype for more complex changes.

```python
# 2.1 .assign() — add column via assign (returns new DF)
print("=== df.assign(NewCol=...) ===")
df_assigned = df.assign(
    TaxRate=0.08,
    Tax=lambda x: x['TotalAmount'] * x['TaxRate'],
    FinalPrice=lambda x: x['TotalAmount'] + x['Tax']
)
print(df_assigned[['Product', 'TotalAmount', 'Tax', 'FinalPrice']].head())
```

```python
# 2.2 .astype() — change dtypes
print("\n=== df.astype({...}) ===")
df_types = df.astype({'Quantity': float, 'Rating': str})
print(df_types[['Quantity', 'Rating']].dtypes)
```

## Task 3: GroupBy Basics
Group data and compute aggregations.

```python
# 3.1 .groupby().sum()
print("=== groupby('Category').sum() ===")
cat_sum = df.groupby('Category')['TotalAmount'].sum()
print(cat_sum)
```

```python
# 3.2 .groupby().mean()
print("\n=== groupby('Category').mean() ===")
cat_mean = df.groupby('Category')['TotalAmount'].mean()
print(cat_mean)
```

```python
# 3.3 .groupby().count()
print("\n=== groupby('Category').count() ===")
cat_count = df.groupby('Category')['Product'].count()
print(cat_count)
```

```python
# 3.4 .groupby().size()
print("\n=== groupby('Category').size() ===")
cat_size = df.groupby('Category').size()
print(cat_size)
```

```python
# 3.5 .groupby().first() and .last()
print("\n=== groupby('Category').first() ===")
cat_first = df.groupby('Category').first()
print(cat_first[['Product', 'TotalAmount']])
```

```python
# 3.6 .groupby().agg() with dict
print("\n=== groupby('Category').agg({...}) ===")
cat_agg = df.groupby('Category').agg({
    'TotalAmount': ['sum', 'mean', 'count'],
    'Quantity': 'mean',
    'Rating': 'max'
})
print(cat_agg)
```

## Task 4: Advanced GroupBy
Pivot tables and groupby-filter chains.

```python
# 4.1 .pivot_table()
print("=== pivot_table ===")
pivot = df.pivot_table(
    values='TotalAmount',
    index='Category',
    columns='PaymentMethod',
    aggfunc='mean',
    fill_value=0
)
print(pivot)
```

```python
# 4.2 GroupBy → Filter chain
print("\n=== groupby → filter: categories with avg > $150 ===")
result = (df.groupby('Category')['TotalAmount']
          .mean()
          .loc[lambda x: x > 150]
          .reset_index())
print(result)
```
