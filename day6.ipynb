# 📅 Day 6 — Missing Data, Apply & Output
**Goal:** Use every missing data handling, apply, and output function at least once.
**Dataset:** `sales_data.csv`

```python
import pandas as pd
import numpy as np

df = pd.read_csv('sales_data.csv')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## Task 1: Detecting Missing Data
Find NaN values in every column.

```python
# 1.1 .isna() — boolean mask of NaN
print("=== df.isna() (first 5 rows) ===")
print(df.isna().head())
```

```python
# 1.2 .isnull() — same as isna()
print("\n=== df.isnull().sum() ===")
print(df.isnull().sum())
```

```python
# 1.3 .notna() — boolean mask of non-NaN
print("\n=== df.notna().sum() ===")
print(df.notna().sum())
```

```python
# 1.4 .notnull() — same as notna()
print("\n=== df.notnull().sum() ===")
print(df.notnull().sum())
```

## Task 2: Handling Missing Data
Introduce NaN values and handle them with every method.

```python
# Introduce some NaN values for practice
df_test = df.copy()
df_test.loc[0, 'Discount'] = np.nan
df_test.loc[1, 'Rating'] = np.nan
df_test.loc[2, 'City'] = np.nan

print("=== NaN counts after introducing ===")
print(df_test[['Discount', 'Rating', 'City']].isna().sum())
```

```python
# 2.1 .dropna() — drop rows with any NaN
print("\n=== df.dropna() ===")
df_dropped = df_test.dropna()
print(f"Rows before: {len(df_test)}, after: {len(df_dropped)}")
```

```python
# 2.2 .dropna(subset=[...]) — drop rows where specific cols are NaN
print("\n=== df.dropna(subset=['Discount']) ===")
df_subset_dropped = df_test.dropna(subset=['Discount'])
print(f"Rows before: {len(df_test)}, after: {len(df_subset_dropped)}")
```

```python
# 2.3 .fillna(value) — fill NaN with a scalar
print("\n=== df.fillna(0) ===")
df_filled_scalar = df_test.fillna(0)
print(f"NaN remaining: {df_filled_scalar.isna().sum().sum()}")
```

```python
# 2.4 .fillna({col: value}) — fill different cols differently
print("\n=== df.fillna({'Discount': 0, 'Rating': 3}) ===")
df_filled_dict = df_test.fillna({'Discount': 0, 'Rating': 3})
print(df_filled_dict[['Discount', 'Rating']].head())
```

```python
# 2.5 .interpolate() — linear interpolation
print("\n=== df.interpolate() ===")
df_interp = df_test.copy()
df_interp['Rating'] = df_interp['Rating'].interpolate(method='linear')
print(df_interp[['Rating']].head(10))
```

```python
# 2.6 Fill with column mean
print("\n=== df['Discount'].fillna(df['Discount'].mean()) ===")
df_mean_filled = df_test.copy()
df_mean_filled['Discount'] = df_mean_filled['Discount'].fillna(df_mean_filled['Discount'].mean())
print(f"NaN remaining: {df_mean_filled['Discount'].isna().sum()}")
```

```python
# 2.7 Fill with column median
print("\n=== df['Rating'].fillna(df['Rating'].median()) ===")
df_median_filled = df_test.copy()
df_median_filled['Rating'] = df_median_filled['Rating'].fillna(df_median_filled['Rating'].median())
print(f"NaN remaining: {df_median_filled['Rating'].isna().sum()}")
```

## Task 3: Apply Functions
Apply custom functions to Series, rows, and DataFrames.

```python
# 3.1 .apply(lambda) on Series — element-wise
print("=== df['TotalAmount'].apply(lambda x: x * 0.9) (first 5) ===")
result = df['TotalAmount'].apply(lambda x: round(x * 0.9, 2))
print(result.head())
```

```python
# 3.2 .apply(lambda, axis=1) — row-wise
print("\n=== df.apply(lambda row: ..., axis=1) ===")
df['Tax'] = df.apply(lambda row: round(row['TotalAmount'] * 0.08, 2), axis=1)
print(df[['Product', 'TotalAmount', 'Tax']].head())
```

```python
# 3.3 .applymap() — element-wise on DataFrame
print("\n=== df[['Quantity', 'Rating']].applymap(lambda x: x * 2) ===")
result = df[['Quantity', 'Rating']].applymap(lambda x: x * 2 if pd.notna(x) else np.nan)
print(result.head())
```

## Task 4: Deduplication & Value Counts
Remove duplicates and count unique values.

```python
# 4.1 .drop_duplicates() — remove duplicate rows
print("=== df.drop_duplicates() ===")
df_no_dupes = df.drop_duplicates()
print(f"Rows before: {len(df)}, after: {len(df_no_dupes)}")
```

```python
# 4.2 .drop_duplicates(subset=[...]) — remove dupes based on specific cols
print("\n=== df.drop_duplicates(subset=['Category']) ===")
df_unique_cat = df.drop_duplicates(subset=['Category'])
print(df_unique_cat[['Category', 'Product']])
```

```python
# 4.3 .value_counts() — count of each unique value
print("\n=== df['Category'].value_counts() ===")
print(df['Category'].value_counts())
```

```python
# 4.4 .nunique() — number of unique values
print("\n=== df['Category'].nunique() ===")
print(f"Unique categories: {df['Category'].nunique()}")
```

## Task 5: Writing Output
Save data to every supported format.

```python
# 5.1 .to_csv()
df.to_csv('output_sales.csv', index=False)
print("✅ Saved to output_sales.csv")
```

```python
# 5.2 .to_excel()
df.to_excel('output_sales.xlsx', index=False)
print("✅ Saved to output_sales.xlsx")
```

```python
# 5.3 .to_json()
df.to_json('output_sales.json', orient='records', indent=2)
print("✅ Saved to output_sales.json")
```

```python
# 5.4 .to_parquet()
df.to_parquet('output_sales.parquet', index=False)
print("✅ Saved to output_sales.parquet")
```
