# Practice Task: Industrial Machine Efficiency Analysis

Welcome to your first practice challenge! This task tests your skills in **Pandas**, **vectorized calculations**, **filtering**, **visualization**, and **data exporting**.

### Scenario
You have been handed a small dataset representing five machines on a production floor, their target outputs, and their actual daily outputs.

### Your Instructions:
1. **Load/Create the DataFrame**: Use the dictionary provided in the first code cell to create a Pandas DataFrame.
2. **Calculate Efficiency**: Create a new column named `Efficiency` using the formula:  
   $$\text{Efficiency (\%)} = \frac{\text{Actual Output}}{\text{Target Output}} \times 100$$
3. **Filter Underperforming Assets**: Identify machines with an `Efficiency` below **85%**. Store them in a new DataFrame called `df_low_efficiency`.
4. **Visualize**: Generate a bar chart showing the efficiency of all machines using Matplotlib or Seaborn. Draw a red horizontal line at the 85% threshold.
5. **Export**: Export the underperforming machines (`df_low_efficiency`) to an Excel sheet named `Underperforming_Machines.xlsx`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Input Dataset
data = {
    "Machine_ID": ["CNC-01", "CNC-02", "Press-01", "Lathe-03", "Oven-01"],
    "Target_Output": [500, 450, 600, 350, 200],
    "Actual_Output": [480, 360, 590, 280, 195]
}

# Start your solution below:
```

```python
# 1. Load/Create the DataFrame
df = pd.DataFrame(data)

# 2. Calculate Efficiency (%)
df["Efficiency"] = (df["Actual_Output"] / df["Target_Output"]) * 100

# 3. Filter Underperforming Assets (< 85%)
df_low_efficiency = df[df["Efficiency"] < 85.0]

# 4. Visualize the Results
plt.figure(figsize=(8, 5))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Machine_ID", y="Efficiency", data=df, hue="Machine_ID", palette="coolwarm", legend=False)

# Draw the 85% threshold line
plt.axhline(y=85.0, color="red", linestyle="--", linewidth=2, label="85% Warning Limit")

# Polish the visualization
plt.title("Production Line Efficiency Diagnostics", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Machine Identifier", fontsize=12)
plt.ylabel("Calculated Efficiency (%)", fontsize=12)
plt.ylim(0, 110)
plt.legend(loc="lower left")
plt.tight_layout()
plt.show()

# 5. Export to Excel Report
output_file = "Underperforming_Machines.xlsx"
df_low_efficiency.to_excel(output_file, index=False)

print(f"--- PIPELINE RUN COMPLETE ---")
print(f"Underperforming machines exported to {output_file}:")
print(df_low_efficiency)
```
