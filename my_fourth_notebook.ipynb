```python

import matplotlib.pyplot as plt
# The X-Axis: 8 hours in a shift
shift_hours = [1, 2, 3, 4, 5, 6, 7, 8]

# The Y-Axis: The average cycle time (in seconds) for each hour
cycle_times = [45, 46, 44, 48, 52, 55, 59, 62]
plt.plot(shift_hours, cycle_times)
plt.title("CNC Machine Cycle Time Trend")
plt.xlabel("Shift Hours")
plt.ylabel("Cycle Time (Seconds)")
plt.axhline(y=50, color='green', linestyle='--', alpha=0.9, label='Normal < 50')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# The X-Axis: Conveyor speed multiplier (1.0 is normal speed, 2.5 is extremely fast)
conveyor_speed = [1.0, 1.2, 1.5, 1.8, 2.0, 2.2, 2.5]

# The Y-Axis: The defect rate (Percentage of broken parts)
defect_rate = [0.5, 0.6, 1.1, 1.4, 1.8, 2.5, 3.2]
plt.scatter(conveyor_speed, defect_rate, color = ['red'], marker = 'x')
plt.title("Impact of Conveyor Speed on Defects")
plt.xlabel("Speed Multiplier")
plt.ylabel("Defect Rate (%)")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- DATASETS ---
machines = ["Press-01", "Lathe-02", "CNC-03", "Oven-04"]
downtime_hours = [2.5, 1.0, 4.0, 0.5]

conveyor_speed = [1.0, 1.2, 1.5, 1.8, 2.0, 2.2, 2.5]
defect_rate = [0.5, 0.6, 1.1, 1.4, 1.8, 2.5, 3.2]

# --- SETUP THE GRID (1 Row, 2 Columns) ---
# figsize controls the size of the overall window width and height
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# --- LEFT PLOT: DOWNTIME BAR CHART ---
# Target the first slot using axs[0]
axs[0].bar(machines, downtime_hours, color='orange')
axs[0].set_title("Downtime by Machine")
axs[0].set_xlabel("Equipment ID")
axs[0].set_ylabel("Downtime (Hours)")

# --- RIGHT PLOT: SCATTER CORRELATION ---
# Task: Target the second slot using axs[1] and write the code to plot your scatter plot
# Use conveyor_speed on X, defect_rate on Y, red color, and marker='x'
# Hint: In subplots, use .set_title(), .set_xlabel(), and .set_ylabel() instead of plt.title()
# WRITE YOUR axs[1] CODE HERE:
axs[1].scatter(conveyor_speed, defect_rate, color='red', marker='x')
axs[1].set_title("CORRELATION")
axs[1].set_xlabel("Conveyor Speed")
axs[1].set_ylabel("Defect Rate")

# --- CLEANUP AND RENDER ---
plt.tight_layout() # Prevents titles and labels from overlapping
plt.show()
```
