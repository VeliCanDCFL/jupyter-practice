## 🧠 Concept: What is Pandas?

Pandas is a Python library for **working with tabular data** — think Excel spreadsheets, CSV files, SQL tables. It gives you two main tools:

| Tool | What it is | Analogy |
|------|-----------|---------|
| **Series** | A single column of data (one-dimensional) | A single column in Excel |
| **DataFrame** | Multiple columns of data (two-dimensional) | An entire Excel sheet |

Think of a DataFrame like this:
```
     Name      Age     City
0   Alice      25      NYC
1     Bob      30      LA
2   Carol      35  Chicago
```

The numbers on the left (0, 1, 2) are called the **index** — they're row labels.
The names on top (Name, Age, City) are **column names**.

---

## 📖 Part 1: Creating a DataFrame from Scratch

### Method A: From a Dictionary of Lists

The most common way to create a DataFrame manually. Each key becomes a column name, each list becomes the column data.

```python
pd.DataFrame({
    'column_name': [value1, value2, value3],
    'another_col': [val1, val2, val3]
})
```

### Method B: From a List of Lists

Each inner list is a ROW. You must specify column names separately.

```python
pd.DataFrame(
    [[row1_col1, row1_col2], [row2_col1, row2_col2]],
    columns=['ColA', 'ColB']
)
```

---

## 📖 Part 2: Loading Data from Files

Pandas can read many file formats. The most common:

| Function | File Type |
|----------|----------|
| `pd.read_csv()` | CSV files (most common) |
| `pd.read_excel()` | Excel (.xlsx) files |
| `pd.read_json()` | JSON files |

---

## 📖 Part 3: Inspecting Your Data

When you load data, the FIRST thing you should do is inspect it. Here's your inspection toolkit:

| Function | What it tells you |
|----------|------------------|
| `.head(n)` | First n rows (default 5) |
| `.tail(n)` | Last n rows (default 5) |
| `.info()` | Column names, types, non-null counts |
| `.describe()` | Statistics for numeric columns (mean, min, max, etc.) |
| `.shape` | (rows, columns) as a tuple |
| `.columns` | List of column names |
| `.dtypes` | Data type of each column |
| `.sample(n)` | Random n rows |

## ⚠️ Common Mistakes

| Mistake | Why it's wrong | Fix |
|---------|---------------|-----|
| Forgetting `pd.` prefix | Pandas isn't imported as just `DataFrame` | Always use `pd.DataFrame()` |
| Mixing up rows vs columns in list method | Inner lists are ROWS, not columns | Each inner list = one row |
| Not checking `.dtypes` | You might think a column is numbers when it's text | Always check dtypes before doing math! |

````python
# ============================================
# BONUS: Create DataFrame from Excel
# ============================================

# First, create a small DF and save it as Excel
sample_df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Carol'],
    'Score': [95, 87, 92],
    'Grade': ['A', 'B+', 'A-']
})
sample_df.to_excel('student_grades.xlsx', index=False)
print("✅ Created student_grades.xlsx")

# Now read it back
df_excel = pd.read_excel('student_grades.xlsx')
print("\n=== Read from Excel ===")
print(df_excel)
````
