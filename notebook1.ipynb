```python
from typing import List, Tuple

def calculate_yield(produced: int, defective: int) -> float:
    """Calculate the yield percentage of a production batch.
    
    Args:
        produced: Number of units produced
        defective: Number of defective units
    
    Returns:
        Yield percentage (0-100)
    """
    if produced <= 0:
        return 0.0 
    return ((produced - defective) / produced) * 100

# Test data - Production Line
products: List[str] = ["Product X", "Product Y", "Product Z", "Product W"]
produced: List[int] = [450, 380, 520, 290]
defective: List[int] = [12, 25, 8, 18]

# Calculate yields
yields = [calculate_yield(p, d) for p, d in zip (produced, defective)]

# Display results
print("Production Yield Analysis: ")
for p, y in zip(products, yields):
    print(f"{p:12} : {y:6.2f}%")
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(products, yields, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, 
             f'{height:.1f}%', ha='center', fontsize=12)

plt.title('Production Yield by Product', fontsize=14, pad=20)
plt.ylabel('Yield (%)')
plt.ylim(0, 105)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
from typing import List, Dict

# List of machines
machines: List[str] = ["CNC-01", "Lathe-02", "Press-03", "Robot-04"]

# Dictionary with machine data
machine_data: Dict[str, Dict] = {
    "CNC-01": {"status": "running", "efficiency": 92, "daily_output": 450},
    "Lathe-02": {"status": "maintenance", "efficiency": 65, "daily_output": 180},
    "Press-03": {"status": "running", "efficiency": 88, "daily_output": 520},
    "Robot-04": {"status": "running", "efficiency": 95, "daily_output": 380}
}
```

```python
def analyze_production_line(machine_data: Dict) -> None:
    """Analyze production line and give IE recommendations."""
    running_count = 0
    total_efficiency = 0
    low_efficiency_machines = []
    
    print("=== Production Line Analysis ===\n")
    
    for machine, info in machine_data.items():
        status = info["status"]
        efficiency = info["efficiency"]
        
        print(f"{machine:10} | Status: {status:12} | Efficiency: {efficiency}%")
        
        if status == "running":
            running_count += 1
            total_efficiency += efficiency
        if efficiency < 75:
            low_efficiency_machines.append(machine)
    
    avg_efficiency = total_efficiency / running_count if running_count > 0 else 0
    
    print(f"\nSummary:")
    print(f"Running machines : {running_count}/{len(machine_data)}")
    print(f"Average efficiency: {avg_efficiency:.1f}%")
    
    if low_efficiency_machines:
        print(f"⚠️  Needs attention : {low_efficiency_machines}")
    else:
        print("✅ All machines performing well!")

# Run the analysis
analyze_production_line(machine_data)
```

```python
from typing import Dict, List
import matplotlib.pyplot as plt

def quality_control_report(product_data: Dict[str, Dict]) -> None:
    """Generate quality control analysis report with summary statistics."""
    print("=== Quality Control Analysis ===\n")
    
    total_defect_rate = 0.0
    critical_count = 0
    
    for product, info in product_data.items():
        defect_rate = info["defect_rate"]
        total_defect_rate += defect_rate
        
        if defect_rate < 2.0:
            status = "Excellent"
        elif defect_rate <= 5.0:
            status = "Acceptable"
        else:
            status = "Critical"
            critical_count += 1
        
        print(f"{product:12} | Defect Rate: {defect_rate:5.2f}% | {status}")
    
    # Summary Statistics
    avg_defect_rate = total_defect_rate / len(product_data)
    
    print("\n" + "="*50)
    print("SUMMARY")
    print(f"Total Products     : {len(product_data)}")
    print(f"Average Defect Rate: {avg_defect_rate:.2f}%")
    print(f"Critical Products  : {critical_count}")
    print("="*50)


# Test Data
product_data: Dict[str, Dict] = {
    "ProductA": {"defect_rate": 5.4},
    "ProductB": {"defect_rate": 4.0},
    "ProductC": {"defect_rate": 3.0},
    "ProductD": {"defect_rate": 0.2},
    "ProductE": {"defect_rate": 1.4},
    "ProductF": {"defect_rate": 6.5}
}

quality_control_report(product_data)
```

```python
# ==================== Visualization ====================

products: List[str] = list(product_data.keys())
defect_rates: List[float] = [info["defect_rate"] for info in product_data.values()]

plt.figure(figsize=(10, 6))
bars = plt.bar(products, defect_rates, color=['red' if x > 5 else 'orange' if x > 2 else 'green' for x in defect_rates])

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.2, 
             f'{height:.1f}%', ha='center', fontsize=11)

plt.title('Product Defect Rates - Quality Analysis', fontsize=14, pad=20)
plt.ylabel('Defect Rate (%)')
plt.ylim(0, max(defect_rates) + 3)
plt.axhline(y=2, color='green', linestyle='--', alpha=0.7, label='Excellent < 2%')
plt.axhline(y=5, color='red', linestyle='--', alpha=0.7, label='Critical > 5%')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
from typing import Dict
import matplotlib.pyplot as plt

def inventory_analyzer(inventory: Dict[str, Dict]) -> None:
    """Analyze inventory status and create visualization."""
    need_reorder = 0
    total_stock = 0.0
    
    print("=== Inventory Analysis Report ===\n")
    
    # Analysis
    for item, info in inventory.items():
        stock = info["stock"]
        min_required = info["min_required"]
        total_stock += stock
        
        if stock < min_required:
            status = "Critical"
            need_reorder += 1
        elif stock < min_required * 1.5:
            status = "Low"
        else:
            status = "Healthy"
        
        print(f"{item:10} | Stock: {stock:3} | Min Required: {min_required:3} | {status}")
    
    avg_stock = total_stock / len(inventory)
    
    print("\n" + "="*55)
    print("SUMMARY")
    print(f"Total Items              : {len(inventory)}")
    print(f"Average Stock Level      : {avg_stock:.1f}")
    print(f"Items Need Reorder       : {need_reorder}")
    print("="*55)

inventory = {
"Item-101": {"stock": 45, "min_required": 50},
"Item-102": {"stock": 120, "min_required": 80},
"Item-103": {"stock": 12, "min_required": 30},
"Item-104": {"stock": 75, "min_required": 60},
"Item-105": {"stock": 8,  "min_required": 25}
}


def plot_inventory(inventory: Dict[str, Dict]) -> None:
    """Create bar chart comparing current stock vs minimum required."""
    items = list(inventory.keys())
    stock = [info["stock"] for info in inventory.values()]
    min_req = [info["min_required"] for info in inventory.values()]
    
    x = range(len(items))   # positions on x-axis
    
    plt.figure(figsize=(12, 7))
    
    plt.bar(x, stock, width=0.4, label='Current Stock', color='skyblue', alpha=0.9)
    plt.bar([i + 0.4 for i in x], min_req, width=0.4, label='Minimum Required', color='orange', alpha=0.7)
    
    plt.xlabel('Items')
    plt.ylabel('Units')
    plt.title('Inventory Status - Current Stock vs Minimum Required')
    plt.xticks([i + 0.2 for i in x], items, rotation=45)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    
    # Add value labels
    for i in range(len(items)):
        plt.text(i, stock[i] + 2, str(stock[i]), ha='center')
        plt.text(i + 0.4, min_req[i] + 2, str(min_req[i]), ha='center')
    
    plt.tight_layout()
    plt.show()






inventory_analyzer(inventory)
plot_inventory(inventory)
```
