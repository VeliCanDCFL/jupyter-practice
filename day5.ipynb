# 📅 Day 5 — Strings & Dates
**Goal:** Use every string and date/time function at least once.
**Dataset:** `sales_data.csv`

```python
import pandas as pd

df = pd.read_csv('sales_data.csv')
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
```

## Task 1: String Operations
Use every `.str.*` method on the CustomerName and Product columns.

```python
# 1.1 .str.upper() — convert to uppercase
print("=== df['CustomerName'].str.upper() (first 5) ===")
print(df['CustomerName'].str.upper().head())
```

```python
# 1.2 .str.lower() — convert to lowercase
print("\n=== df['Product'].str.lower() (first 5) ===")
print(df['Product'].str.lower().head())
```

```python
# 1.3 .str.strip() — remove whitespace
print("\n=== df['City'].str.strip() (first 5) ===")
print(df['City'].str.strip().head())
```

```python
# 1.4 .str.contains() — check substring
print("\n=== df['Product'].str.contains('Laptop|Phone|Tablet') ===")
electronics_like = df[df['Product'].str.contains('Laptop|Phone|Tablet', case=False)]
print(f"Found {len(electronics_like)} matching products")
print(electronics_like[['Product']].head())
```

```python
# 1.5 .str.replace() — replace substring
print("\n=== df['Product'].str.replace(' ', '_') (first 5) ===")
print(df['Product'].str.replace(' ', '_').head())
```

```python
# 1.6 .str.split() — split into lists
print("\n=== df['CustomerName'].str.split() (first 5) ===")
print(df['CustomerName'].str.split().head())
```

```python
# 1.7 .str.len() — string length
print("\n=== df['Product'].str.len() (first 5) ===")
print(df['Product'].str.len().head())
```

```python
# 1.8 .str.extract() — regex extraction
print("\n=== df['Comments'].str.extract(r'(\\d+)' ) (first 5) ===")
extracted = df['Comments'].str.extract(r'(\d+)')
print(extracted.head())
```

## Task 2: Date/Time Operations
Use every `.dt.*` method on the OrderDate column.

```python
# 2.1 .dt.year
print("=== df['OrderDate'].dt.year (first 5) ===")
print(df['OrderDate'].dt.year.head())
```

```python
# 2.2 .dt.month
print("\n=== df['OrderDate'].dt.month (first 5) ===")
print(df['OrderDate'].dt.month.head())
```

```python
# 2.3 .dt.day
print("\n=== df['OrderDate'].dt.day (first 5) ===")
print(df['OrderDate'].dt.day.head())
```

```python
# 2.4 .dt.hour (create a time column for demo)
print("\n=== df['OrderDate'].dt.hour ===")
df_demo = pd.read_csv('sales_data.csv')
df_demo['OrderDateTime'] = pd.to_datetime(
    df_demo['OrderDate'] + ' 10:30:00'
)
print(df_demo['OrderDateTime'].dt.hour.head())
```

```python
# 2.5 .dt.minute
print("\n=== df['OrderDateTime'].dt.minute ===")
print(df_demo['OrderDateTime'].dt.minute.head())
```

```python
# 2.6 .dt.second
print("\n=== df['OrderDateTime'].dt.second ===")
print(df_demo['OrderDateTime'].dt.second.head())
```

```python
# 2.7 .dt.dayofweek (0=Monday, 6=Sunday)
print("\n=== df['OrderDate'].dt.dayofweek (first 5) ===")
print(df['OrderDate'].dt.dayofweek.head())
```

```python
# 2.8 .dt.day_name()
print("\n=== df['OrderDate'].dt.day_name() (first 5) ===")
print(df['OrderDate'].dt.day_name().head())
```

```python
# 2.9 .dt.month_name()
print("\n=== df['OrderDate'].dt.month_name() (first 5) ===")
print(df['OrderDate'].dt.month_name().head())
```

```python
# 2.10 .to_period('M') — convert to period
print("\n=== df['OrderDate'].dt.to_period('M') (first 5) ===")
print(df['OrderDate'].dt.to_period('M').head())
```

```python
# 2.11 pd.Grouper(freq='D') — group by day
print("\n=== pd.Grouper(freq='D') in groupby ===")
df_daily = df.set_index('OrderDate').groupby(pd.Grouper(freq='D')).agg({
    'TotalAmount': 'sum',
    'TransactionID': 'count'
}).head(10)
print(df_daily)
```

## Task 3: Combined String + Date Pipeline
Clean text and extract date info in one chain.

```python
# Clean customer names: strip whitespace, uppercase
df['CleanName'] = df['CustomerName'].str.strip().str.upper()

# Extract first name and last name
df['FirstName'] = df['CustomerName'].str.split().str[0]
df['LastName'] = df['CustomerName'].str.split().str[-1]

# Add date components
df['OrderYear'] = df['OrderDate'].dt.year
df['OrderMonth'] = df['OrderDate'].dt.month_name()
df['OrderDayOfWeek'] = df['OrderDate'].dt.day_name()

print("=== Combined pipeline result (first 5) ===")
print(df[['CustomerName', 'CleanName', 'FirstName', 'LastName', 
          'OrderYear', 'OrderMonth', 'OrderDayOfWeek']].head())
```
