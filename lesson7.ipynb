## 🎯 MISSION 1: Load & Understand

Load the data and inspect it thoroughly. Document what you find.

```python
# ============================================
# PHASE 1: Load & Understand
# ============================================

import pandas as pd

# Load data
df = pd.read_csv('sales_data.csv')

# Inspect
print("=== BASIC INFO ===")
print(f"Shape: {df.shape}")
print(f"Columns ({len(df.columns)}): {df.columns.tolist()}")
print(f"\nDtypes:\n{df.dtypes}")

print("\n=== FIRST 5 ROWS ===")
print(df.head())

print("\n=== LAST 3 ROWS ===")
print(df.tail(3))

print("\n=== DESCRIBE (numeric) ===")
print(df.describe())

print("\n=== MISSING VALUES ===")
print(df.isna().sum())
```

---

## 🎯 MISSION 2: Clean the Data

Fix all issues you found. Document each cleaning step.

```python
# ============================================
# PHASE 2: Clean the Data
# ============================================

# Step 1: Convert OrderDate to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print("✅ Converted OrderDate to datetime")

# Step 2: Handle missing Discount (fill with 0)
df['Discount'] = df['Discount'].fillna(0)
print("✅ Filled missing Discount with 0")

# Step 3: Handle missing Rating (fill with median)
median_rating = df['Rating'].median()
df['Rating'] = df['Rating'].fillna(median_rating)
print(f"✅ Filled missing Rating with median ({median_rating})")

# Step 4: Handle missing City (fill with 'Unknown')
df['City'] = df['City'].fillna('Unknown')
print("✅ Filled missing City with 'Unknown'")

# Step 5: Drop pending orders (incomplete data)
df_clean = df[df['Status'] == 'Completed'].copy()
print(f"✅ Dropped pending orders. Remaining: {len(df_clean)} rows")

# Step 6: Remove duplicates
before = len(df_clean)
df_clean = df_clean.drop_duplicates(subset=['TransactionID'])
print(f"✅ Removed {before - len(df_clean)} duplicates")

# Step 7: Standardize text
df_clean['Category'] = df_clean['Category'].str.strip().str.title()
df_clean['City'] = df_clean['City'].str.strip().str.title()
print("✅ Standardized text (strip + title case)")

# Verify no NaNs remain
print(f"\n=== Final NaN count: {df_clean.isna().sum().sum()} ===")
```

---

## 🎯 MISSION 3: Explore & Analyze

Answer each of your boss's questions with code. Show the data that supports your answer.

```python
# ============================================
# PHASE 3: Explore & Analyze
# ============================================

# === QUESTION 1: Top-selling categories? ===
print("=== Q1: Revenue by Category ===")
cat_revenue = df_clean.groupby('Category')['TotalAmount'].agg(['sum', 'mean', 'count'])
cat_revenue.columns = ['TotalRevenue', 'AvgOrderValue', 'OrderCount']
cat_revenue = cat_revenue.sort_values('TotalRevenue', ascending=False)
print(cat_revenue)

# === QUESTION 2: Which city spends the most? ===
print("\n=== Q2: Revenue by City ===")
city_revenue = df_clean.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)
print(city_revenue.head(10))

# === QUESTION 3: Average order value by payment method? ===
print("\n=== Q3: Avg Order Value by Payment Method ===")
payment_avg = df_clean.groupby('PaymentMethod')['TotalAmount'].agg(['mean', 'count'])
payment_avg.columns = ['AvgValue', 'OrderCount']
print(payment_avg)

# === QUESTION 4: Products with highest ratings? ===
print("\n=== Q4: Top Rated Products ===")
product_rating = df_clean.groupby('Product')['Rating'].mean().sort_values(ascending=False)
print(product_rating.head(10))

# === QUESTION 5: Sales by day of week? ===
print("\n=== Q5: Sales by Day of Week ===")
df_clean['DayName'] = df_clean['OrderDate'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_sales = df_clean.groupby('DayName')['TotalAmount'].sum()
day_sales = day_sales.reindex(day_order)  # Keep correct order
print(day_sales)
```

---

## 🎯 MISSION 4: Create New Insights

Go beyond the basics. Add columns that reveal hidden patterns.

```python
# ============================================
# PHASE 4: Create New Insights
# ============================================

# Insight 1: Calculate discount savings
df_clean['DiscountSaved'] = df_clean['TotalAmount'] * df_clean['Discount']
print("=== Total Discount Savings ===")
print(f"${df_clean['DiscountSaved'].sum():,.2f} saved across all orders")

# Insight 2: Add tax column
df_clean['Tax'] = df_clean['TotalAmount'] * 0.08
print(f"\n=== Total Tax Collected ===")
print(f"${df_clean['Tax'].sum():,.2f}")

# Insight 3: Extract date components
df_clean['Year'] = df_clean['OrderDate'].dt.year
df_clean['Month'] = df_clean['OrderDate'].dt.month_name()
df_clean['Quarter'] = df_clean['OrderDate'].dt.quarter
print("\n=== Sales by Month ===")
monthly = df_clean.groupby('Month')['TotalAmount'].sum()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
print(monthly.reindex(month_order).dropna())

# Insight 4: Customer spending segments
customer_spending = df_clean.groupby('CustomerName')['TotalAmount'].sum().reset_index()
customer_spending['Segment'] = pd.cut(
    customer_spending['TotalAmount'],
    bins=[0, 100, 300, 1000],
    labels=['Low', 'Medium', 'High']
)
print("\n=== Customer Spending Segments ===")
print(customer_spending['Segment'].value_counts())

# Insight 5: Price range analysis
df_clean['PriceRange'] = pd.cut(
    df_clean['TotalAmount'],
    bins=[0, 50, 150, 300, 500, 1000],
    labels=['$0-50', '$51-150', '$151-300', '$301-500', '$501+']
)
print("\n=== Orders by Price Range ===")
print(df_clean['PriceRange'].value_counts().sort_index())
```

---

## 🎯 MISSION 5: Present Results

Create a summary report and save your work.

```python
# ============================================
# PHASE 5: Present Results
# ============================================

# Create a summary DataFrame
summary = pd.DataFrame({
    'Metric': [
        'Total Revenue',
        'Average Order Value',
        'Total Orders',
        'Total Tax',
        'Total Discount Given',
        'Unique Customers',
        'Unique Categories',
        'Avg Rating'
    ],
    'Value': [
        f"${df_clean['TotalAmount'].sum():,.2f}",
        f"${df_clean['TotalAmount'].mean():,.2f}",
        f"{len(df_clean):,}",
        f"${df_clean['Tax'].sum():,.2f}",
        f"${df_clean['DiscountSaved'].sum():,.2f}",
        f"{df_clean['CustomerName'].nunique()}",
        f"{df_clean['Category'].nunique()}",
        f"{df_clean['Rating'].mean():.1f}/5"
    ]
})

print("=" * 50)
print("📊 SALES ANALYSIS REPORT")
print("=" * 50)
for _, row in summary.iterrows():
    print(f"{row['Metric']:.<35} {row['Value']}")
print("=" * 50)
```

```python
# Save your work
df_clean.to_csv('analysis_output.csv', index=False)
print("✅ Saved cleaned data to analysis_output.csv")

# Save summary as Excel
summary.to_excel('analysis_summary.xlsx', index=False)
print("✅ Saved summary to analysis_summary.xlsx")
```
