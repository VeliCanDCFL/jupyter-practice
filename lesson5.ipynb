## 🧠 Concept: Working with Text and Dates

Real-world data is messy. You'll often need to:
1. **Clean text** — fix capitalization, remove spaces, extract info
2. **Parse dates** — convert strings to proper date objects
3. **Extract date parts** — get year, month, day of week from a date

---

## 📖 Part 1: String Operations (`.str.*`)

To use string methods, you access the `.str` accessor on a Series:

```python
df['Name'].str.upper()        # Convert to uppercase
df['Name'].str.lower()        # Convert to lowercase
df['Name'].str.strip()        # Remove leading/trailing whitespace
df['Name'].str.contains('pat') # Check if text contains a pattern
df['Name'].str.replace('a','b') # Replace text
df['Name'].str.split(',')     # Split text into lists
df['Name'].str.len()          # Length of each string
df['Name'].str.extract(r'(\d+)') # Extract with regex
```

**Key insight:** Every `.str.*` method returns a Series of the same length — one result per row.

---

## 📖 Part 2: Date Operations (`.dt.*`)

First, convert strings to dates with `pd.to_datetime()`:

```python
df['DateCol'] = pd.to_datetime(df['DateCol'])
```

Then use the `.dt` accessor to extract parts:

```python
df['DateCol'].dt.year         # Year (2024)
df['DateCol'].dt.month        # Month (1-12)
df['DateCol'].dt.day          # Day of month (1-31)
df['DateCol'].dt.hour         # Hour (0-23)
df['DateCol'].dt.minute       # Minute (0-59)
df['DateCol'].dt.second       # Second (0-59)
df['DateCol'].dt.dayofweek    # Day of week (0=Monday, 6=Sunday)
df['DateCol'].dt.day_name()   # Full name ('Monday')
df['DateCol'].dt.month_name() # Full name ('January')
df['DateCol'].dt.to_period('M') # Convert to month period
```

---

## 📖 Part 3: Grouping by Date with `pd.Grouper()`

When your index is a date, you can group by day, week, month, etc.:

```python
df.set_index('OrderDate').groupby(pd.Grouper(freq='D')).sum()  # Daily
df.set_index('OrderDate').groupby(pd.Grouper(freq='W')).sum()  # Weekly
df.set_index('OrderDate').groupby(pd.Grouper(freq='M')).sum()  # Monthly
```

1. **`.str.*` methods clean text** — uppercase, lowercase, strip, split, replace, extract

````python
# ============================================
# MISSION 1: String Operations
# ============================================

import pandas as pd

df = pd.read_csv('sales_data.csv')
print(f"Loaded {len(df)} rows")

# 1. Uppercase customer names
print("=== Uppercase Names ===")
df['NameUpper'] = df['CustomerName'].str.upper()
print(df[['CustomerName', 'NameUpper']].head())

# 2. Lowercase products
print("\n=== Lowercase Products ===")
df['ProductLower'] = df['Product'].str.lower()
print(df[['Product', 'ProductLower']].head())

# 3. Strip whitespace from City
print("\n=== Stripped City ===")
df['CityClean'] = df['City'].str.strip()
print(df[['City', 'CityClean']].head())

# 4. Contains with regex (Laptop OR Phone)
print("\n=== Products containing Laptop or Phone ===")
tech_products = df[df['Product'].str.contains('Laptop|Phone', case=False)]
print(f"Found {len(tech_products)} products")
print(tech_products[['Product']].head())

# 5. Replace spaces with underscores
print("\n=== Replaced Spaces ===")
df['ProductClean'] = df['Product'].str.replace(' ', '_')
print(df[['Product', 'ProductClean']].head())

# 6. Split name into first and last
print("\n=== Split Names ===")
df['FirstName'] = df['CustomerName'].str.split().str[0]
df['LastName'] = df['CustomerName'].str.split().str[-1]
print(df[['CustomerName', 'FirstName', 'LastName']].head())

# 7. Length of product name
print("\n=== Product Name Lengths ===")
df['ProductNameLen'] = df['Product'].str.len()
print(df[['Product', 'ProductNameLen']].head())
````

````python
# ============================================
# MISSION 2: Date Operations
# ============================================

# Convert to datetime first!
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Extract date components
df['Year'] = df['OrderDate'].dt.year
df['Month'] = df['OrderDate'].dt.month
df['Day'] = df['OrderDate'].dt.day
df['DayOfWeek'] = df['OrderDate'].dt.dayofweek
df['DayName'] = df['OrderDate'].dt.day_name()
df['MonthName'] = df['OrderDate'].dt.month_name()
df['Period'] = df['OrderDate'].dt.to_period('M')

print("=== Date Components ===")
print(df[['Product', 'OrderDate', 'Year', 'Month', 'Day', 'DayOfWeek', 'DayName', 'MonthName', 'Period']].head())
````

````python
# ============================================
# MISSION 3: Group by Date
# ============================================

# Set OrderDate as index
df_date = df.set_index('OrderDate')

# Daily aggregation
print("=== Daily Sales (first 5 days) ===")
daily = df_date.groupby(pd.Grouper(freq='D')).agg({
    'TotalAmount': 'sum',
    'TransactionID': 'count'
}).head(5)
print(daily)

# Monthly aggregation
print("\n=== Monthly Orders ===")
monthly = df_date.groupby(pd.Grouper(freq='M')).agg({
    'TotalAmount': ['sum', 'mean'],
    'TransactionID': 'count'
})
print(monthly)

# Best and worst sales days
best_day = daily['TotalAmount'].idxmax()
worst_day = daily['TotalAmount'].idxmin()
print(f"\nBest sales day: {best_day} (${daily.loc[best_day, 'TotalAmount']:,.2f})")
print(f"Worst sales day: {worst_day} (${daily.loc[worst_day, 'TotalAmount']:,.2f})")
````

````python
# ============================================
# BONUS: Clean a Messy Text Column
# ============================================

# Simulate messy data
messy = pd.DataFrame({
    'RawEmail': ['  Alice@Example.COM ', 'bob@test.org', 'CAROL@TEST.COM', 'david@example.com'],
    'RawName': ['  alice johnson  ', ' BOB SMITH ', 'carol white', 'David Brown']
})

print("=== Before Cleaning ===")
print(messy)

# Clean email: strip + lowercase
messy['CleanEmail'] = messy['RawEmail'].str.strip().str.lower()

# Clean name: strip + title case (first letter capitalized)
messy['CleanName'] = messy['RawName'].str.strip().str.title()

print("\n=== After Cleaning ===")
print(messy[['RawEmail', 'CleanEmail', 'RawName', 'CleanName']])
````
